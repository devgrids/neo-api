```python
from gtts import gTTS

# Convertir texto a habla y guardarla en un archivo de audio
text = "Hola, cómo estás?, me llamo Yordy"
tts = gTTS(text=text, lang='es')
tts.save("salida.mp3")
```

```python
from transformers import pipeline
from pydub import AudioSegment
import numpy as np

# Crear el pipeline
pipe = pipeline("text-to-speech", model="facebook/mms-tts-eng")
# pipe = pipeline("text-to-speech", model="facebook/mms-tts-spa")

# pipe = pipeline("text-to-speech", model="suno/bark")

# Definir el texto que deseas convertir a habla
text = "Hello, my name is Yordy"

# Generar el habla
output = pipe(text)

# Convertir los datos de audio al formato adecuado para PyDub
audio_data = np.array(output['audio']).astype(np.float32)
audio_data /= np.max(np.abs(audio_data))
audio_data = (audio_data * 32767).astype(np.int16)

# Crear el objeto AudioSegment
audio = AudioSegment(audio_data.tobytes(), frame_rate=22050, sample_width=2, channels=1)

# Guardar el habla como un archivo MP3
audio.export("output.mp3", format="mp3")
```
